```python
def render_word_html(tok: str, rgba: tuple):
    if (rgba[0] + rgba[1] + rgba[2]) / 3 < 0.5: text_color = "white"
    else: text_color = "black"

    return f"<span style=\"color: {text_color}; background-color: rgb({rgba[0] * 255}, {rgba[1] * 255}, {rgba[2] * 255})\">{tok} </span>"


def render_word_latex(tok: str, rgba: tuple):
    if (rgba[0] + rgba[1] + rgba[2]) / 3 < 0.5: pre_text, post_text = "{\\color{" + "white}", "}"
    else: pre_text, post_text = "{\\color{" + "black}", "}"
    # pre_text, post_text = "", ""

    tok = unidecode.unidecode(tok.replace("_", "\_").replace("$", "\$"))
    return pre_text + "\\ctext[RGB]{"+f"{rgba[0] * 255},{rgba[1] * 255},{rgba[2] * 255}"+ "}{" + tok + " }" + post_text


def visualize_explain(text, label):
    gradient, tokens, _ = explain(text, label)
    cmap = sns.color_palette("viridis", as_cmap=True)
    normalizer = matplotlib.colors.Normalize(vmin=0., vmax=gradient[:-1].mean().item() + gradient[:-1].std().item() * 2)

    html_string = ""
    latex_string = ""
    for grad, tok in zip(gradient, tokens):
        html_string += render_word_html(tok, cmap(normalizer(grad.item())))
        latex_string += render_word_latex(tok, cmap(normalizer(grad.item())))

    display(HTML(html_string))
    print(latex_string)
```
